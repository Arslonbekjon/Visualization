```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
df =df[df.year>2000]
plt.figure(figsize=(10,6))

sns.scatterplot(data=df,x='mileage',y='price')
#sns.regplot(data=df,x='mileage',y='price')
plt.show()
```

```python
df =df[df.year>2000]
plt.figure(figsize=(10,6))

sns.scatterplot(data=df,x='mileage',y='price',hue="transmission",s=100)
#sns.regplot(data=df,x='mileage',y='price')
plt.show()
```

```python
df =df[df.year>2000]
plt.figure(figsize=(10,6))

sns.scatterplot(data=df,x='mileage',y='price',hue="transmission",style='fuelType',s=100)
#sns.regplot(data=df,x='mileage',y='price')
plt.show()
```

```python
df =df[df.year>2000]
plt.figure(figsize=(10,6))

sns.scatterplot(data=df,x='mileage',y='price',hue="fuelType",s=100)
#sns.regplot(data=df,x='mileage',y='price')
plt.show()
```
