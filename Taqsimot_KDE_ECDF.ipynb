```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df = df[df.year>=2010]
df = df[df.transmission!="Other"]
df = df[df.fuelType!="Other"]
df.head()
```

# KDE plot

```python
sns.displot(data=df,x='price',kind='kde',height=5,aspect=2)
plt.show()
```

```python
sns.displot(data=df,x='price',kde=True,height=5,aspect=2)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.kdeplot(data=df,x='price',fill=True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.kdeplot(data=df,x='price',hue='fuelType',fill=True)
plt.show()
```

```python
sns.displot(data=df,x='price',hue='fuelType',kind='kde',height=5,aspect=2,fill=True)
plt.xticks(rotation=90)
plt.show()
```

# ECDF plot

```python
sns.displot(data=df,x='price',kind="ecdf",height=5,aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both') # setka major minor both qiymatlarini berib ko'rish kerak
plt.minorticks_on()
plt.show()
```
